```python
chems = ['BENZENE', 'CREOSOTE', 'POLYCHLORINATED BIPHENYLS']
chems_dfs = [df_benzene, df_creosote, df_pcb]
for year in range(1987,2017):
    for chem, df_chem in zip(chems, chems_dfs):
        fate_list = []
        fate_list = [chem, year, df_chem.iloc[0]['UNIT_OF_MEASURE']]
        for category in fates[3:70]: #['5.1_FUGITIVE_AIR', '5.2_STACK_AIR', '5.3_WATER']:
            fate = df_chem[(df_chem['YEAR']==year) & (df_chem[category])][category].sum()
            fate_list.append(fate)
        row = pd.Series(fate_list, df_Fates.columns)
        df_Fates=df_Fates.append(row, ignore_index=True)            
```
